```python
def fibonacci(n: int) -> list[int]:
  """
  Retourne la suite de fibonacci contenant n termes
  """
  #initialisation
  if n < 0:
    raise ValueError(" n doit etre positif ou null")
  if n == 0:
    return []
  if n == 1:
    return [0]

  fib = [0,1]
  for _ in range(2, n):
    #Ajouter la liste d'element forme par la somme des deux precedents
    fib.append(fib[-1] + fib[-2])
  return fib
```

```python
def fibonacci_gen(n: int):
  if n <= 0:
    return
  a, b = 0, 1
  for _ in range(n):
    yield a
    a, b = b, a+b

for f in fibonacci_gen(100000):
  if f > 1000:
    print('Stop a:', f)
    break
```

```python
def pascal(n: int)->list[int]:
  if n < 1:
    raise ValueError(" n must be >= 1")
  if n == 1:
    return [1]

  previous = pascal(n-1)#get the previous row
  middle = [previous[i] + previous[i+1] for i in range(len(previous)-1)]
  return [1] + previous + [1]
```

```python
def pascal_iter(n: int)-> list[int]:
  if n < 1:
    raise ValueError('n must be >= 1')
  row = [1]
  for _ in range(2,n+1):
    row = [1] + [row[i] + row[i+1] for i in range(len(row) -1)] + [1]
  return row
```

```python
def uv(n: int):
  if n < 0:
    raise v(" n must be >=0")
  u, v = 1, 1
  for _ in range(n):
    u, v = u + v, 2*u -v
  return u, v
print(uv(100))
```

```python
def vk(n0, K):
  """
  Compute K steps of the collatz sequence starting from n0
  and return the last 5 values.
  """
  if K <1 :
    raise ValueError(' k must be >= 1')
  v = int(n0)
  res = [] #liste to store the last 5 result

  for k in range(K):
    if v % 2 == 0:
      v //= 2
    else:
      v = 3 * v + 1

    if k >= K-5:
      res.append(v)

  return res

for n0 in [10, 100, 1000, 10000]:
  print(vk(n0, 1000))
```

```python
def position(mots, x, n):
  return [mot for mot in mots if n<len(mot) and mot[n]==x]


mots = ['Abricot', 'Airelle', 'Ananas', 'Banane', 'Cassis', 'Cerise', 'Citron', 'Clémentine', 'Coing', 'Datte', 'Fraise', 'Framboise', 'Grenade', 'Groseille', 'Kaki', 'Kiwi', 'Litchi', 'Mandarine', 'Mangue', 'Melon', 'Mirabelle', 'Nectarine', 'Orange', 'Pamplemousse', 'Papaye', 'Pêche', 'Poire', 'Pomme', 'Prune', 'Raisin']
print(position(mots,'e', 4))
```

```python
from collections import defaultdict

mots_dict =  defaultdict(list)

for word in mots:
  for i, ch in enumerate(word):
    mots_dict[(ch,i)].append(word)

print(mots_dict[('e',4)])
```
